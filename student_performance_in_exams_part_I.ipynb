This notebook explores a dataset containing certain scores and attributes for 1000 students, including their `gender`, `race/ethnicity`, `lunch`, and `test preparation course`. the dataset can be found in this Kaggle notebook [here.] (https://www.kaggle.com/datasets/spscientist/students-performance-in-exams)

```python
# import all packages and set plots to be embedded inline with certain appearance changes
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

sb.set_style("darkgrid", {'xtick.bottom': True, 'ytick.left': True})
sb.set_context("paper", font_scale=1.4)
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('StudentsPerformance.csv')

# high-level overview of data shape and composition
print(df.shape)
df.head(10)
```

```python
df.info()
```

```python
# renaming columns for ease of use
df.rename(columns = {'race/ethnicity' : 'ethnicity'}, inplace = True)
df.columns = df.columns.str.replace(" ", "_")
df.head()
```

```python
# print statistics
df.describe()
```

```python
# creating a column for percentage
df["percentage"] = round((df.math_score + df.reading_score + df.writing_score)/3, 2)
df.head(2)
```

###### We will create a grading criteria based on percentage and then put it in column

```python
# function for grading criteria based on percentage
def grade(score):
    x = int(10 - score/10) 
    if x >= 6:
        x = 6
    return ['A', 'B+', 'B', 'C', 'D', 'E', 'F'][x]
```

Grading criteria:

- 91-100 - A
- 81-90 - B+
- 71-80 - B 
- 61-70 - C 
- 51-60 - D 
- 41-50 - E 
- Less than or equal to 40 - F

```python
# creating grade columns
df['grade'] = df.percentage.apply(grade)
df.head(2)
```

```python
# convert grade column into categorical column
grade_order = pd.api.types.CategoricalDtype(categories = ['A', 'B+', 'B', 'C', 'D', 'E', 'F'], ordered = True)
df.grade = df.grade.astype(grade_order)
```
